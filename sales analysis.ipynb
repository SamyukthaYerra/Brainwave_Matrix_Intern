```python
import matplotlib.pyplot as plt

# Step 1: Group and summarize
sales_per_Category = df.groupby('Category')['Total Sale'].sum().reset_index()
sales_per_Category = sales_per_Category.rename(columns={'Category': 'Customer Type', 'Total Sale': 'Total Sales'})

# Step 2: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    sales_per_Category['Total Sales'],
    labels=sales_per_Category['Customer Type'],
    autopct='%1.1f%%',
    startangle=140
)

plt.title('Total Sales Distribution per Customer Category')
plt.axis('equal')  # Keeps pie chart circular
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Group and summarize
sales_per_Category = df.groupby('Category')['Total Sale'].sum().reset_index()
sales_per_Category = sales_per_Category.rename(columns={'Category': 'Customer Type', 'Total Sale': 'Total Sales'})

# Step 2: Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(
    sales_per_Category['Customer Type'],
    sales_per_Category['Total Sales'],
    color='skyblue'
)

# Add labels and title
plt.title('Total Sales per Customer Category')
plt.xlabel('Customer Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=30)  # Rotate x-axis labels if needed
plt.tight_layout()
plt.show()
```

```python
#convert order date to date time format

df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)

#grouping by year and summing the sales per year

yearly_sales = df.groupby(df["Date"].dt.year)["Total Sale"].sum()

#setting new index and renaming the columns
yearly_sales = yearly_sales.reset_index()
yearly_sales = yearly_sales.rename(columns={"Sales": "Total Sale"})

print(yearly_sales)
```

```python
yearly_sales = df.groupby(df['Date'].dt.year)['Total Sale'].sum().reset_index()
yearly_sales = yearly_sales.rename(columns={'Date': 'Year'})  # if it's named 'Date'
# OR (more likely)
yearly_sales = yearly_sales.rename(columns={yearly_sales.columns[0]:"Year"})
```

```python

plt.bar(yearly_sales['Year'], yearly_sales['Total Sale'])
plt.xlabel("Year")
plt.ylabel("Total Sale")
plt.title("Yearly Sales")
plt.show()


```
